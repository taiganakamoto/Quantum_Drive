今井さんの定式化で考察してみる（gが小さい領域で有効な解析）
gが小さい領域では、電場への反作用が無視でき、$\hat{a} \rightarrow \alpha e^{i\omega_c t}$の時間発展が起きる。
そうすると、干渉項と非干渉項のダイナミクスを独立に考えることができて、最後にその寄与を足せば良くなる。
$$
\hat{\rho}_e(t) = \rho_1(t) + \rho_2(t) + e^{-2|\alpha|^2}(\rho_3(t) + \rho_4(t)) \\
i\partial_t \rho_i(t) = [H_i(t),\rho_i(t)] \\
H_1(t) = H_e + 2g \alpha_0 \cos(\omega_c t) \\
H_2(t) = H_e - 2g \alpha_0 \cos(\omega_c t) \\
H_3(t) = H_e + 2ig \alpha_0 \sin(\omega_c t) \\
H_4(t) = H_e - 2ig \alpha_0 \sin(\omega_c t)
$$
ハミルトニアンの1と2には振動電場に相当する場がかかっている。これはx方向と-x方向に電場がかかっていることに対応するので、今の場合電流に対する寄与は（空間反転対称性より）0である。
また、ハミルトニアンの3と4には複素振動電場に相当する場がかかっている。ただし、これもy方向と-y方向に電場がかかるので、今の場合に電流に対する寄与は（空間反転対称性より）0である。
複素の場がかかっているので、時間反転対称性は破れていそうなので、そこで非自明な寄与が取り出せれば面白いと思うが、状態全体をみると時間反転対称性は破れていないので、難しいのか？
